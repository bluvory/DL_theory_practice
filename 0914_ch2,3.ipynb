```python
import numpy as np

def OR(x1,x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.1
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
def AND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
def NAND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
# non-linear
# multi-layer

def XOR(x1,x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    
    return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

## Step function (Heaviside Funtion)
- small change in weight or bias can cause rapid change in output !

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    if x>0:
        return 1
    else:
        return 0
    
x = np.linspace(-5, 5, 101)
y = np.zeros(101,)

for i in range(0,101):
    y[i] = step_function(x[i])
    
plt.plot(x,y)
plt.show()
```

```python
def step_function(x):
    y = x > 0
    return y.astype(np.int)

x = np.linspace(-5, 5, 101)
y = step_function(x)

plt.plot(x,y)
plt.show()
```

## Activation Function
- Remedy: Smooting out near discontinuity !
- We call H(x) as an activation function
- Finding good activation functions -> research field !
- Let us consider some well-known activation functions

```python
# Sigmoid function

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-5, 5, 101)
y = sigmoid(x)

plt.plot(x,y)
plt.show()
```

```python
# Arctan function

def arctan(x):
    return np.arctan(x)

x = np.linspace(-5, 5, 101)
y = arctan(x)

plt.plot(x,y)
plt.show()
```

```python
# ReLU function

def relu(x):
    return np.maximum(x,0)

x = np.linspace(-5, 5, 101)
y = relu(x)

plt.plot(x,y)
plt.show()
```
